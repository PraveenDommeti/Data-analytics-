```python

import pandas as pd

df = pd.read_csv('electric_vehicles_spec_2025.csv.csv')

print("First row:")
print(df.head())

print("\nLast row:")
print(df.tail())
```

```python
print(df.info())

print(df.describe())
```

```python
# Handling Missing Values

# Fill all missing values with 0
df_filled_zero = df.fillna(0)
```

```python

print(df_filled_zero)
```

```python
# Drop rows with missing values

df_dropped_rows = df.dropna()

# Drop columns with missing values
df_dropped_cols = df.dropna(axis=1)
print(df_dropped_cols)
```

```python
sum = df['efficiency_wh_per_km'].sum()
print(sum)

mean = df['top_speed_kmh'].mean()
print(mean)

median = df['fast_charging_power_kw_dc'].median()
print(median)

mode = df['number_of_cells'].mode()
print(mode)
```
